create calculated metrics

```python
def calculate_metrics(df, ground_truth_col, predicted_col):
    def soft_metrics(row, gt_col, pred_col):
        gt_set = set(row[gt_col])
        pred_set = set(row[pred_col])
        intersection = len(gt_set.intersection(pred_set))
        
        precision = intersection / len(pred_set) if len(pred_set) > 0 else 0
        recall = intersection / len(gt_set) if len(gt_set) > 0 else 0
        fscore = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        
        return precision, recall, fscore

    # Apply calculation row-wise
    metrics = df.apply(lambda row: soft_metrics(row, ground_truth_col, predicted_col), axis=1)
    df[f'Soft_Precision_{ground_truth_col}_{predicted_col}'], df[f'Soft_Recall_{ground_truth_col}_{predicted_col}'], df[f'Soft_FScore_{ground_truth_col}_{predicted_col}'] = zip(*metrics)
    
    # Calculate means
    mean_precision = df[f'Soft_Precision_{ground_truth_col}_{predicted_col}'].mean() * 100
    mean_recall = df[f'Soft_Recall_{ground_truth_col}_{predicted_col}'].mean() * 100
    mean_fscore = df[f'Soft_FScore_{ground_truth_col}_{predicted_col}'].mean() * 100
    
    print(f"Mean Soft Precision_{ground_truth_col}_{predicted_col}:", round(mean_precision, 2), "%")
    print(f"Mean Soft Recall_{ground_truth_col}_{predicted_col}:", round(mean_recall, 2), "%")
    print(f"Mean Soft F-Score_{ground_truth_col}_{predicted_col}:", round(mean_fscore, 2), "%")
    
    # return{mean_precision, mean_recall, mean_fscore }
    return {
        'Mean Soft Precision (%)': mean_precision,  
        'Mean Soft Recall (%)': mean_recall,     
        'Mean Soft F-Score (%)': mean_fscore     
    }
```

```python
column_pairs = [
    ('Annotated_CUIs', 'OA_CUIs'),
    ('Annotated_TUIs', 'OA_TUIs'),
    ('Annotated_CUIs', 'cTAKES_CUIs'),
    ('Annotated_TUIs', 'cTAKES_TUIs'),
    ('Annotated_CUIs', 'MM_CUIs'),
    ('Annotated_TUIs', 'MM_TUIs'),
    ('Annotated_CUIs', 'ensemble_1_CUIs'),
    ('Annotated_TUIs', 'ensemble_1_TUIs'),
    ('Annotated_CUIs', 'ensemble_2_CUIs'),
    ('Annotated_TUIs', 'ensemble_2_TUIs'),
    ('Annotated_CUIs', 'ensemble_3_CUIs'),
    ('Annotated_TUIs', 'ensemble_3_TUIs'),
    
]

# Initialize an empty dictionary to store results
results = {}

# Calculate metrics for each pair and collect results in a structured way
for ground_truth, predicted in column_pairs:
    metric_results = calculate_metrics(complete_dataset, ground_truth, predicted)
    for key, value in metric_results.items():
        if key not in results:
            results[key] = []
        results[key].append(value)

# Convert the structured results into a DataFrame
results_df = pd.DataFrame(results, index=[pair[1] for pair in column_pairs])
```

```python
results_df
```

```python
# Map system names
system_map = {
    'OA_CUIs': 'OpenAlex(OA)',
    'OA_TUIs': 'OpenAlex(OA)',
    'cTAKES_CUIs': 'cTAKES',
    'cTAKES_TUIs': 'cTAKES',
    'MM_CUIs': 'MetaMap(MM)',
    'MM_TUIs': 'MetaMap(MM)',
    'ensemble_1_CUIs': 'ensemble_1 (OA+cTAKES)',
    'ensemble_1_TUIs': 'ensemble_1 (OA+cTAKES)',
    'ensemble_2_CUIs': 'ensemble_2 (OA+MM)',
    'ensemble_2_TUIs': 'ensemble_2 (OA+MM)',
    'ensemble_3_CUIs': 'ensemble_3 (OA+cTAKES+MM)',
    'ensemble_3_TUIs': 'ensemble_3 (OA+cTAKES+MM)'
}

# Convert original data to the desired format
transformed_data = {
    'System': [],
    'Soft Recall (CUI)': [],
    'Soft Recall (TUI)': [],
    'Soft Precision (CUI)': [],
    'Soft Precision (TUI)': [],
    'Soft F-Score (CUI)': [],
    'Soft F-Score (TUI)': []
}


for index, row in results_df.iterrows():
    system = system_map[index]
    precision_cui = row['Mean Soft Precision (%)']
    recall_cui = row['Mean Soft Recall (%)']
    f1_cui = row['Mean Soft F-Score (%)']
    
    # Handle TUIs separately
    if '_CUIs' in index:
        transformed_data['System'].append(system)
        transformed_data['Soft Recall (CUI)'].append(recall_cui)
        transformed_data['Soft Precision (CUI)'].append(precision_cui)
        transformed_data['Soft F-Score (CUI)'].append(f1_cui)
    else:
        transformed_data['Soft Recall (TUI)'].append(recall_cui)
        transformed_data['Soft Precision (TUI)'].append(precision_cui)
        transformed_data['Soft F-Score (TUI)'].append(f1_cui)

# Create DataFrame
result_table = pd.DataFrame(transformed_data)
```
